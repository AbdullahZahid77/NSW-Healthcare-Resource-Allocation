```python
import pandas as pd

# 1. Path to your Excel file
file_path = r'C:\Users\abdul\OneDrive\Desktop\sydney_project\BHI_DATA_HQ_ElectiveSurgery_Jan2010–Sep2025.xlsx'

try:
    # 2. If "overall_results" is a SHEET inside the Excel file:
    df_surgery = pd.read_excel(file_path, sheet_name='overall_results')
    

    # Display the first few rows and column info
    print("--- Data Preview ---")
    print(df_surgery.head())
    print("\n--- Column Info ---")
    print(df_surgery.info())

except Exception as e:
    print(f"Error: {e}")
```

```python
# Filter for the most recent quarter
target_quarter = 'Jul-Sep 2025'
df_elective = df_surgery[df_surgery['reporting_period'] == target_quarter]
df_emergency = df_emergency[df_emergency['reporting_period'] == target_quarter]

# Filter for "Hospital" level (ignoring NSW-wide totals for now)
df_elective = df_elective[df_elective['reporting_level'] == 'Hospital']
df_emergency = df_emergency[df_emergency['reporting_level'] == 'Hospital']
```

```python
# Pivot Elective Surgery
elective_pivot = df_elective.pivot_table(
    index=['entity_longname', 'lhd_shortname', 'peer_group_shortname'], 
    columns='measure_name', 
    values='measure_value', 
    aggfunc='first'
).reset_index()

# Pivot Emergency Dept
emergency_pivot = df_emergency.pivot_table(
    index=['entity_longname'], 
    columns='measure_name', 
    values='measure_value', 
    aggfunc='first'
).reset_index()

# Merge the two
combined_df = pd.merge(elective_pivot, emergency_pivot, on='entity_longname', how='inner')
```

```python
combined_df
```

```python
# Convert values to numeric (handling errors for suppressed data like '<5')
cols_to_fix = ['Percentage of elective surgeries performed on time', 'Percentage of patients starting treatment on time']
for col in cols_to_fix:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

# Drop hospitals with missing data for a clean 1-day analysis
combined_df = combined_df.dropna(subset=cols_to_fix)

# Simple Pressure Score: Average of "On-Time Failures"
combined_df['Pressure_Score'] = (
    (100 - combined_df['Percentage of elective surgeries performed on time']) + 
    (100 - combined_df['Percentage of patients starting treatment on time'])
) / 2
```
